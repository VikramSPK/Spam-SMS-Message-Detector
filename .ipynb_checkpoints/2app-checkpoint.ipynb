```python
from flask import Flask,render_template,url_for,request
import pandas as pd 
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import joblib
import re
import string
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
# from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
# import joblib

```

```python
app = Flask(__name__)

@app.route('/')
def home():
	return render_template('home1.html')

@app.route('/predict',methods=['POST'])

def predict():   
    
    df = pd.read_csv('text_data.csv', encoding="ISO-8859-1")
    print(df.head(10))
    df.drop(['S. No.'], axis=1, inplace=True)
    df['label'] = df['Label'].map({'Non-Spam': 0, 'Spam': 1})
    s=pd.Series(df['Message_body'])
    df['Message_body']=s.str.lower()
    df['Message_body'] = df['Message_body'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '' , x))
    df['Message_body']= df['Message_body'].apply(lambda x: re.sub('W*dw*','',x))

    stop_words = set(stopwords.words('english'))
    stop_words.add('subject')
    stop_words.add('http')
    def remove_stopwords(text):
        return " ".join([word for word in str(text).split() if word not in stop_words])
    df['Message_body']=  df['Message_body'].apply(lambda x: remove_stopwords(x))
    # print(df.head(10))
    lemmatizer = WordNetLemmatizer()
    def lemmatize_words(text):
        return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
    df['Message_body']= df['Message_body'].apply(lambda text: lemmatize_words(text))
    # print(df.head(10))
    df['Message_body']= df['Message_body'].apply(lambda text: re.sub(' +', ' ',text))
    print(df.head(10))
    
    X = df['Message_body']
    y = df['Label']
    cv = CountVectorizer()
    X = cv.fit_transform(X) # Fit the Data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    #Naive Bayes Classifier
    clf = MultinomialNB()
    clf.fit(X_train,y_train)
    clf.score(X_test,y_test)
    y_pred = clf.predict(X_test)
#     joblib.dump(clf, 'NB_spam_model.pkl')
    print("Heloow!!")
    if request.method == 'POST':
        message = request.form['message']
        data = [message]
#         data=cv.fit_transform(data)
        vect = cv.transform(data).toarray()
        print(vect)
        my_prediction = clf.predict(vect)
    print(my_prediction)
    if( my_prediction=="Spam"):
         my_prediction=1
    else:
        my_prediction=0

    return render_template('result1.html',prediction = my_prediction)
    
if __name__ == '__main__':
    app.run()
```
